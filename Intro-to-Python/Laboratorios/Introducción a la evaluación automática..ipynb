## Funciones:

Las funciones en python son muy parecidas a las funciones que conocemos por nuestras clases de matemática, donde tenemos una caja negra que toma un valor y nos retorna un resultado. En Python se utilizan funciones para poder almacenar y anidar código, que queremos utilizar de forma recurrente, de una forma amigable y modularizada. 

Las funciones se definene utilizando el comando <span style="color: #008000; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> def </span>, el que es seguido por el nombre de la función, y entre parentesis, la declaración de los parametros que esta toma en caso de recibir algo por el usuario. Al igual que las sentencias if-else, for o while, se termina la linea con un ":".

Por ejemplo, si quisieramos una función llamada suma que reciba dos valores 'a' y 'b', vea si se pueden sumar, y en caso de poderse, nos retorne su resultado, dberíamos tener una función escrita de la siguiente manera:

```python 
def suma(a,b):
    if type(a) == type(b):
            return a + b
    else:
        if type(a) in [int,float, bool]:
            if type(b) in [int,float, bool]:
                return a + b
        else:
            print('Los parametros tienen que ser ambos números o ambos del mismo tipo en otro caso.')
            return None         
```

## Testeo:

El testeo de funciones es básicamente conocer el comportamiento esperado de una funcion y sus casos borde, antes de que esten desarrolladas. De esta forma, sabemos que si una función pasa todos los test a los que se le somete, quiere decir que funciona como se esperaba y no esta dejando de lado los casos borde que estan considerados al momento de planearla.

Para evaluar las funciones que crearás, simplemente las invocaremos con valores especificos y esperaremos que nos retornen el calor correcto y de la manera esperada.

## Importancia del nombre de una función:

Al conocer el nombre que deben tener tus funciones, podemos utilizar código para llamar a las funciones que crearás y podremos testearlas de manera automática con distintos casos, por esto, es necesario que <span style="color: #800000; display:inline-block; font-size:16px; font-weight: bold; font-style: oblique;"> tengan mucho ojo con que el nombre de la función sea el que les pedimos</span>

### Return v/s Print

Dentro de Python existe un comando muy importante al utilizar funciones, este es el <span style="color: #008000; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> return </span>, el cual se utiliza para que el programa pueda volver a utilizar los valores que entrega la función y es diferente a lo que hace print.

#### Return:
Comando que debemos utilizar al finalizar una función que queremos que nos entregue un valor para utilizar este valor.

#### Print:
Comando que nos permite imprimir lo que le pongamos como parametros a imprimir, nos ayuda a visualizar información o estados de datos, pero no nos devuelve esos valores a alguien fuera de la función.

Si definimos suma como:
```python 
def suma(a,b):
    if type(a) == type(b):
            return a + b
    else:
        if type(a) in [int,float, bool]:
            if type(b) in [int,float, bool]:
                return a + b
        else:
            print('Los parametros tienen que ser ambos números o ambos del mismo tipo en otro caso.')
            return None
```

Al hacer algo del tipo

```python
a = suma(1,3)
```

Tendremos que la variable <span style="color: #003FFF; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> a </span> tiene guardado al numero 4, el que fue retornado por la función suma. Pero por otro lado:

Si definimos suma como:
```python 
def suma(a,b):
    if type(a) == type(b):
            print(a + b)
    else:
        if type(a) in [int,float, bool]:
            if type(b) in [int,float, bool]:
                print(a + b)
        else:
            print('Los parametros tienen que ser ambos números o ambos del mismo tipo en otro caso.')
```

Al hacer algo del tipo

```python
a = suma(1,3)
```
Aparecerá el resultado 4 en la pantalla, pero la variable <span style="color: #003FFF; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> a </span> tendrá almacenado el valor None (valor que usa python cuando no existe un valor almacenado) ya que la función no le retorna nada que almacenar al no tener el comando <span style="color: #008000; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> return </span> cuando quiere que finalice la ejecución de la función.

Cuando una función ejecuta el <span style="color: #008000; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> return </span>, automaticamente deja de ejecutar más cosas dentro de la función.

<span style="color: #008000; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> return </span> es un comando de salida automático de las funciones, por lo que si existe código debajo de la linea del <span style="color: #008000; display:inline-block; font-size:15px; font-weight: bold; font-style: oblique;"> return </span> y en la misma indentación o en otras más la derecha en el mismo camino, ese codigo nunca se ejecutará.
